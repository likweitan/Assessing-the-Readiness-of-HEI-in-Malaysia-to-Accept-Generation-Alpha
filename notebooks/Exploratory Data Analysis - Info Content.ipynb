```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O
import matplotlib.pyplot as plt
```

```python
%%time
#info_content_raw_df = pd.read_csv('data/Info_Content.csv', index_col='ucid')
info_userdata_raw_df = pd.read_csv('data/Info_UserData.csv', index_col='uuid')
log_problem_raw_df = pd.read_csv('data/Log_Problem.csv', index_col='upid')
```

```python
# scale down some records
log_problem_df = log_problem_raw_df[log_problem_raw_df['timestamp_TW'] < '2019-01-01']
```

```python
log_problem_df
```

```python
#Creating User Table with sums of data points
users = log_problem_df.groupby('uuid').sum()

user_table = pd.DataFrame(users)
```

```python
user_table
```

```python
log_problem_df['timestamp_TW'] = pd.to_datetime(log_problem_df['timestamp_TW'], format='%Y-%m-%d %H:%M:%S %Z')
```

```python
#Creating a time variable rng
user_time = log_problem_df.groupby('uuid')
rng = (user_time['timestamp_TW'].max() - user_time['timestamp_TW'].min()).dt.days
```

```python
active_from_last= (log_problem_df['timestamp_TW'].max() - (user_time['timestamp_TW'].max())).dt.days
user_table['active_from_last'] = active_from_last
```
